# Looping through dictionary

```python
Fruits_List=["apple","bannana","cherry"]
for fruit in Fruits_List:
    print(fruit)
```

```python
student={"name":"Osandha","age":15,"grade":"A"}
for key in student:
    print(key,":",student[key])
```

```python
for letter in "Hello":
    print(letter)
```

```python
for letter in "Python":
    print(letter,end=" ")
```

# While loop

```python
i=0
while i<10:
    print(i)
    i+=1
```

```python
password=""
while password != "python123":
    password=input("Enter your password: ")
print("Access granted")
```

```python
while True:
    command=input("Enter 'exit' to quit: ")
    if command=="exit":
        break
```

```python
battery= 10
while battery>0:
    print("Battery level: ",battery)
    battery-=1
print("Battery empty")
```

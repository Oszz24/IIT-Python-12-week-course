```python
Count=True
while Count:
    UserNum=int(input("Enter a number: "))
    if UserNum%2==0:
        print("Even number")
    else:
        print("Odd number")
    UserCont=input("Do you want to enter another number: ")
    if UserCont=="No" or UserCont=="no":
        Count=False
    else:
        Count=True
```

```python
UserNum=int(input("Enter your marks: "))
if UserNum>=0 and UserNum<=100:
    if UserNum>=90:
        print("A")
    elif UserNum>=80 and UserNum<=89:
        print("B")
    elif UserNum>=70 and UserNum<=79:
        print("C")
    elif UserNum>=60 and UserNum<=69:
        print("D")
    else:
        print("Fail")
else:
    print("Invalid score entered")
```
